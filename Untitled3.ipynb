```python
import numpy as np
import matplotlib.pyplot as plt

def make_quad(a, b, c):
    def f(x):
        return a * x ** 2 + b * x + c
    return f
f = make_quad(a=1,b=2,c=3)
x = np.linspace(-10, 10, 100)

plt.plot(x, f(x))
plt.show()
```
