```python
import pandas as pd
```

```python
df = pd.DataFrame({
                      "name": ['apple','banana','carrot','durian','eggplant'],
                     "price" : [1000,2000,3000,4000,5000],
                     "size" : [10,20,30,40,45],
                     "category" : ['fruit','vegetable','fruit','vegetable','fruit']
                     })
```

```python
print(df["price"].mean())
print(df["price"].median())
print(df["price"].min())
print(df["price"].max())
print(df["price"].std())
print(df["price"].var())
print(df["price"].count())
print(df["price"].sum())
```

```python
print(df["price"].quantile(0.25))
print(df["price"].quantile(0.5))
print(df["price"].quantile(0.75))
print(df["price"].corr(df["size"]))
```
